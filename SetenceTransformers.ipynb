**Generate Embeddings**

```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
sentences = ["Sentence 1", "Setence 2", "Sentence Embeddings"]

embeddings = model.encode(sentences)
print(embeddings)
```

**Cosine Similarity**

```python
emb1 = model.encode("Computers are a huge revolution")
emb2 = model.encode("I like pandas")

cos_sim = util.cos_sim(emb1, emb2)
print("Cosine Similarity:", cos_sim)
```

**Cosine Siilarity between all pairs**

```python
sentences = ["Sentence 1", "Setence 2", "Sentence Embeddings", "I like pandas", "Computers are a huge revolution"]

embeddings = model.encode(sentences)
cos_sim = util.cos_sim(embeddings, embeddings)
print(cos_sim)
```

```python
all_sentence_combinations = []
for i in range(len(cos_sim)-1):
  for j in range(i+1, len(cos_sim)):
    all_sentence_combinations.append((cos_sim[i][j], i, j))

print(all_sentence_combinations)
```

```python
all_sentence_combinations = sorted(all_sentence_combinations, key=lambda x: x[0], reverse=True)
print(all_sentence_combinations)

print("Top 3 similar pairs")
for score, i, j in all_sentence_combinations[0:3]:
  print("{} \t {} \t {:.4f}".format(sentences[i], sentences[j], cos_sim[i][j]))
```

**Semantic Search**

```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
new_question = "What are the primary colors?"
long_answer1 = "The primary colors in the additive color model (like for screens) are red, green, and blue. In the subtractive color model (like for pigments), they are typically considered to be red, yellow, and blue, although cyan, magenta, and yellow are also often used, especially in printing."
long_answer2 = "The average lifespan of a common housefly is about 15 to 30 days. This can vary depending on environmental conditions such as temperature and food availability."
long_answer3 = "Mount Everest, the world's highest mountain above sea level, is located in the Mahalangur Himal sub-range of the Himalayas. Its summit is 8,848.86 meters (29,031.7 feet) above sea level."

query_embeddings = model.encode(new_question)
corpus_embeddings = model.encode([long_answer1, long_answer2, long_answer3])

print(util.semantic_search(query_embeddings, corpus_embeddings)) # Gives the Document
```

**Clustering**

```python
from sklearn.cluster import KMeans
import numpy as np

embedder = SentenceTransformer('all-MiniLM-L6-v2')

corpus = [
    """
    The primary colors in the additive color model (like for screens) are red, green, and blue.
    In the subtractive color model (like for pigments), they are typically considered to be red, yellow, and blue,
    although cyan, magenta, and yellow are also often used, especially in printing.
    """,
    """
    The average lifespan of a common housefly is about 15 to 30 days.
    This can vary depending on environmental conditions such as temperature and food availability.
    """,
    """
    Mount Everest, the world's highest mountain above sea level, is located in the Mahalangur Himal sub-range of the Himalayas.
    Its summit is 8,848.86 meters (29,031.7 feet) above sea level.
    """,
    """
    This is a long piece of text that provides some background information.
    It discusses various topics and sets the scene.
    The capital of Australia is Canberra, a planned city known for its geometric design and national institutions.
    More unrelated information follows here.
    This text continues to provide additional details and context.
    """,
    """
    The Amazon rainforest is the largest rainforest in the world,
    covering an area of 5.5 million square kilometers.
    It is home to an incredible diversity of plant and animal life,
    many of which are unique to the region.
    Deforestation is a major threat to the Amazon,
    driven by factors such as cattle ranching and agriculture.

    In other news, the annual migration of wildebeest in the Serengeti is a spectacular natural event.
    Millions of wildebeest, zebras, and gazelles move in a massive herd across the plains
    in search of fresh grazing and water. This migration is fraught with danger,
    as predators like lions and crocodiles lie in wait.

    Meanwhile, scientists have made a breakthrough in developing a new type of solar panel
    that is significantly more efficient at converting sunlight into electricity.
    This could revolutionize renewable energy production and help combat climate change.
    The research is still in its early stages, but the initial results are promising.

    Furthermore, the history of pizza is a fascinating journey that begins in Naples, Italy.
    Originally a simple flatbread with toppings, pizza evolved into the iconic dish we know today.
    Different regions of Italy and the world have developed their own unique styles and variations.
    """
]

corpus_embeddings = embedder.encode(corpus)

#Normalise the embeddings to unit length
corpus_embeddings = corpus_embeddings / np.linalg.norm(corpus_embeddings, axis=1, keepdims=True)
```

```python
clustering_model = KMeans(n_clusters = 3)
clustering_model.fit(corpus_embeddings)
cluster_assignment = clustering_model.labels_

print(cluster_assignment)
```

```python
clustering_sentence = {}
for sentence_id, cluster_id in enumerate(cluster_assignment):
  if cluster_id not in clustering_sentence:
    clustering_sentence[cluster_id] = []

  clustering_sentence[cluster_id].append(corpus[sentence_id])

clustering_sentence
```
